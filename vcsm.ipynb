## Preprocesamiento de los datos 

```python
#leemos el dataset 
df = pd.read_csv("coches.csv",index_col = 'company')
```

## Relacion de variables test de Pearson

```python
df.corr(method = 'pearson').sort_values(by = 'is_professional', axis = 0).sort_values(by = 'is_professional', axis = 1)
```

```python
corr = df.corr(method = 'pearson').sort_values(by = 'is_professional', axis = 0, ascending = False).sort_values(by = 'is_professional', axis = 1, ascending = False)
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))
 
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(8,8))
 
# Generate a custom diverging colormap  
cmap = sns.diverging_palette(150, 275, l=55, as_cmap=True)
 
# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr.iloc[0:,0:], mask=mask[0:,0:], cmap=cmap, vmax=1, center=0,
            square=True, linewidths=3, cbar_kws={"shrink": .8}, annot = True)
```
